**Example 1.2 - 1 [Computing linear transformations]**

```python
import pandas as pd

gradesPath = 'DataSets/Grades.txt'
columns = ['Phisics','Maths','French','History','Literature']
grades_df = pd.read_csv(gradesPath)
grades_df
```

```python
import numpy as np
np.matrix(grades_df[columns].values)
```

```python
np.matrix([[1/5,1/5,1/5,1/5,1/5],
         [1/2,1/2,0,0,0],
         [0,0,1/3,1/3,1/3]]).transpose()
```

```python
import numpy as np
gradesM = np.matrix(grades_df[columns].values)
matrixT = np.matrix([[1/5,1/5,1/5,1/5,1/5],
                     [1/2,1/2,0,0,0],
                     [0,0,1/3,1/3,1/3]]).transpose()

transformedDataMatrix = np.dot(gradesM,matrixT)
print('Transformation matrix\n',matrixT)

pd.DataFrame(transformedDataMatrix,columns=['OverallMean','ScienceMean','LettersMean'])
```

**Example 1.2 - 1 [Standarizing the average grades DF and computing overall variability measures ]**

Standardizing the average grades of a subset of columns (Phisics,Maths,Literature) avoiding the cross join
___

```python
def getStdDataMatrix(data_df,columns,keys):
    data_df = data_df.copy(deep=True)
    meanVector = data_df[columns].mean().to_dict()
    stdVector = data_df[columns].std().to_dict()
    for col in columns:
        data_df['std_'+col] = (data_df[col]-meanVector[col])/stdVector[col]
    return data_df[keys+['std_'+col for col in columns]]
```

```python
columns = ['Phisics','Maths','Literature']
std_grades_df = getStdDataMatrix(grades_df,columns,keys=['Student'])
std_grades_df
```

```python
covarianceSz = std_grades_df[['std_'+col for col in columns]].cov()
covarianceSz
```

```python
correlationRx = grades_df[columns].corr()
correlationRx
```
